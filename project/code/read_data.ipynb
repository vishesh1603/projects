```python
import json
```

```python
def display_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data = display_data("data.json")
```

```python
def user_data(data):
    for i in data["users"]:
        print(f"ID no. - {i['id']} {i['name']} is friends with {i['friends']} with liked pages - {i['liked_pages']}")
```

```python
user_data(data)
```

```python
def pages_data(data):
    for j in data["pages"]:
        print(F"Page ID {j['id']} belongs to {j['name']}")
```

```python
user_data(data)
pages_data(data)
```

```python
def combined_data(data):
    for user in data["users"]:
        for page in data["pages"]:
            print(f"ID no. - {user['id']} {user['name']} is friends with {[page[user['friends']]} with liked pages - {user['liked_pages']}")
```
