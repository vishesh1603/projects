```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        return json.load(f)
```

```python
def people_u_may_know(user_id,data):
    user_friends = {}
    for user in data["users"]:
        user_friends[user["id"]] = set(user["friends"])
    if user_id not in user_friends:
        return []
    # return user_friends
    # type(user_friends)
    
    direct_friends = user_friends[user_id]

    recc = {}

    for friend in direct_friends:
        if friend not in user_friends:
            continue
        for mutual in user_friends[friend]:
            if mutual != user_id and mutual not in direct_friends:
                recc[mutual] = recc.get(mutual,0) + 1
                
    sorted_suggestions = sorted(recc.items(), key=lambda x: x[1], reverse=True)
    return [user_id for user_id, _ in sorted_suggestions]
```

```python
data = load_data("big_data.json")
user_id = 1  # Example: Finding suggestions for Amit
recommendations = people_u_may_know(user_id, data)
print(f"People You May Know for User {user_id}: {recommendations}")
```
