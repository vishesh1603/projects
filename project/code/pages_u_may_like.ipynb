```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        return json.load(f)
```

```python
def pages_you_might_like(user_id, data):
    user_pages = {}
    for user in data["users"]:
        user_pages[user["id"]] = set(user["liked_pages"])

    if user_id not in user_pages:
        return []

    user_liked_pages = user_pages[user_id]

    recc = {}

    for other_user,pages in user_pages.items():
        if user_id != other_user:
            shared_pages = user_liked_pages.intersection(pages)
            for page in pages:
                if page not in user_liked_pages:
                    recc[page] = recc.get(page , 0) + len(shared_pages)

    sorted_recc = sorted(recc.items(), key = lambda x : x[1], reverse=True)
    return [page_id for page_id, _ in sorted_recc]
```

```python
data = load_data("big_data.json")
user_id = 1
page_recommendations = pages_you_might_like(user_id, data)
print(f"Pages You Might Like for User {user_id}: {page_recommendations}")
```
