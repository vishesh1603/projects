```python
import pandas as pd
```

```python
df = pd.read_csv("extended_dataset.txt")
```

```python
df.dropna(inplace = True)
df.drop_duplicates(inplace = True)
```

```python
df.info
```

```python
df_count = df['race'].value_counts() 
print(df_count)
```

```python
df_male = df[df['sex' ] == 'Male']
df_male['age'].mean()
```

```python
df_bachelors = df[df['education'] == 'Bachelors']
bachelor_percent = int(df_bachelors.shape[0])/int(df.shape[0])*100
print(bachelor_percent)
```

```python
df_advanced = df.query(f"education == 'Bachelors' or education == 'Masters' or education == 'Doctorate'")
df_50k = df_advanced.query(f"salary == '>50K'")
print(df_50k)
```

```python
advanced_moreThan50k_percent = int(df_50k.shape[0])/int(df_advanced.shape[0])*100
print(advanced_moreThan50k_percent)
```

```python
df_notAdvanced = df.query(f"education != 'Bachelors' and education != 'Masters' and education != 'Doctorate'")
df_n50k = df_notAdvanced.query(f"salary == '>50K'")
print(int(df_n50k.shape[0])/int(df_notAdvanced.shape[0])*100)
```

```python
print(df['hours-per-week'].max())
```

```python
min_work = df['hours-per-week'].min()
df_minWork = df[df["hours-per-week"] == min_work]
df_min_50k = df_minWork.query(f"salary == '>50K'")
print(int(df_min_50k.shape[0])/int(df_minWork.shape[0])*100)
```

```python
df_salaryMoreThan50k = df[df['salary'] == '>50K']
print(df_salaryMoreThan50k['native-country'].mode())
df_india_50k = df_salaryMoreThan50k[df_salaryMoreThan50k['native-country'] == 'India']
print(int(df_india_50k.shape[0])/int(df_salaryMoreThan50k.shape[0])*100)
```
